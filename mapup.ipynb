```python
df2 = pd.read_excel('/content/Python&Excel_Assesment.xlsx', "DataSet2")
```

```python
#combining two sheets
value1= df1[['name','Id_a','Id_b','startDay','startTime','endDay','endTime','ablePri2AxlesAuto','ablePri3AxlesAuto','ablePri4AxlesAuto','ablePri2AxlesDualTire','ablePri2AxlesTaxi','ablePri2AxlesTaxiPool','able23AxlesAuto']]
```

```python
value2= df2[['name','Id_a','Id_b','startDay','startTime','endDay','endTime','ablePri2AxlesAuto','ablePri3AxlesAuto','ablePri4AxlesAuto','ablePri2AxlesDualTire','ablePri2AxlesTaxi','ablePri2AxlesTaxiPool','able23AxlesAuto']]
```

```python
dataframes=[value1,value2]
```

```python
join=pd.concat(dataframes)
```

```python
join.shape
```

```python
#replacing null value with specified value
join['ablePri2AxlesAuto'] = join['ablePri2AxlesAuto'].fillna(128)
join['ablePri3AxlesAuto'] = join['ablePri3AxlesAuto'].fillna(64)
join['able23AxlesAuto'] = join['able23AxlesAuto'].fillna(2)
```

```python
join["Id"] = join["Id_a"] + join["Id_b"]
join = join.drop('Id_a', 1)
join = join.drop('Id_b', 1)

```
